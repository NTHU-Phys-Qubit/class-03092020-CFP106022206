```python
import matplotlib.pyplot as plt
import numpy as np
```

## Problem 1

Plot sin(x)

```python
x = np.linspace(0,np.pi**np.pi,10**7)
y = np.sin(x)
plt.style.use("dark_background")
plt.plot(x,y,"-c.")
plt.show()
```
